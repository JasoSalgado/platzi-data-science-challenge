# ¿Qué libro tiene mayor número de reviews con mejor puntuación?

```python
# Having into account that we must obtain the reviews and then, the best rating. I decided to divided the problem
# Firstly, I will get the higher number of reviews

the_most_reviews = df.sort_values('text_reviews_count', ascending = False).head(10).set_index('title')
plt.figure(figsize=(15,10))
sns.barplot(the_most_reviews['text_reviews_count'], the_most_reviews.index, alpha=.9,palette='Set2')
plt.show()
```

```python
# I will get the best rating

the_most_rating = df.sort_values('ratings_count', ascending = False).head(10).set_index('title')
plt.figure(figsize=(15,10))
sns.barplot(the_most_rating['ratings_count'], the_most_rating.index, alpha=.9,palette='Set3')
plt.show()
```

# Mostrar las soluciones con gráficos

```python
# Once, I got the two problems solved.
# I need two variables to get the problem.

reviews = df['text_reviews_count']
ratings = df['ratings_count']>4
both = pd.DataFrame(df[reviews & ratings].sort_values('average_rating', ascending = True).head(10))

plt.figure(figsize = (15,10))
sns.barplot(y = both['authors'], x = both['average_rating'], palette='deep')
plt.show()
```
